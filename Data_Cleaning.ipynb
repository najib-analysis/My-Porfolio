```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"C:\Users\HP\Downloads\Customer Call List.xlsx")
df
```

```python
df=df.drop_duplicates()
df=df.drop(columns="Not_Useful_Column")
df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("..._/")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("-","")
df["Phone_Number"] = df["Phone_Number"].str.replace("|","")
df["Phone_Number"] = df["Phone_Number"].str.replace("/","")
df["Phone_Number"] = df["Phone_Number"].str.replace("N/a","")
df
```

```python
df["Phone_Number"]= df["Phone_Number"].apply(lambda x: str(x))
df
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("nan--","")
df["Phone_Number"] = df["Phone_Number"].str.replace("--","")
df
```

```python
#df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',2, expand=True)
#the above didnt work for me
#s=df["Address"].str.split(pat=" ",n=0, expand = True)
s=df["Address"].str.split(pat=",",n=1, expand = True)
s
```

```python
df["Street_Address"]=s[0]
df["State"]=s[1]
df
```

```python
s=df["Street_Address"].str.split(pat=" ",n=1, expand = True)
s
```

```python
df["Zip_code"]=s[0]
df["Street_Address"]=s[1]
```

```python
df
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace("Yes","Y")
df["Paying Customer"]=df["Paying Customer"].str.replace("No","N")
df["Paying Customer"]=df["Paying Customer"].str.replace("N/a","")
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace("No","N")
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace("Yes","Y")
df
```

```python
df= df.replace("N/a","")
df= df.replace("Na","")
df
```

```python
df=df.fillna("")
```

```python
df
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] =="Y":
        df.drop(x,inplace = True)
df
```

```python
#df=df.dropna(subset="Phone_Number",inplace = True)
#another way to drop null values
for x in df.index:
    if df.loc[x,"Phone_Number"] =="":
        df.drop(x,inplace = True)
df
```
